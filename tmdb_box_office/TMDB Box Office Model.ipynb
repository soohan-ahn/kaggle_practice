```python
import pandas as pd
import numpy as np
from collections import Counter

train_file_name = './tmdb-box-office-prediction/train.csv'
test_file_name = './tmdb-box-office-prediction/test.csv'
train = pd.read_csv(train_file_name)
test = pd.read_csv(test_file_name)
```

```python
print(train.iloc[0])
```

```python
def get_dictionary(x):
    try:
        d = eval(x)
    except:
        d = {}
    return d
```

```python
train.belongs_to_collection = train.belongs_to_collection.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
test.belongs_to_collection = test.belongs_to_collection.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
```

```python
list_of_collections = list(train['belongs_to_collection'].values)
Counter([i for j in list_of_collections for i in j]).most_common()

train.belongs_to_collection = train.belongs_to_collection.map(lambda x: 1 if len(x) > 0 else 0)
test.belongs_to_collection = test.belongs_to_collection.map(lambda x: 1 if len(x) > 0 else 0)
```

```python
train.genres = train.genres.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
test.genres = test.genres.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))

train['genres'].apply(lambda x: len(x) if x != {} else 0).value_counts()
```

```python
for i, e in enumerate(train.genres[:5]):
    print(i, e, len(e))
    
list_of_genres = list(train['genres'].values)

Counter([i for j in list_of_genres for i in j]).most_common()
```

```python
top_genres = [str(m[0]) for m in Counter([i for j in list_of_genres for i in j]).most_common(15)]
for g in top_genres:
    train['genres_' + g] = train.genres.apply(lambda x: 1 if int(g) in x else 0)
    test['genres_' + g] = test.genres.apply(lambda x: 1 if int(g) in x else 0)

train['num_of_genres'] = train.genres.map(lambda x: len(x))
test['num_of_genres'] = test.genres.map(lambda x: len(x))
```

```python
train.production_companies = train.production_companies.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
test.production_companies = test.production_companies.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))

train['production_companies'].apply(lambda x: len(x) if x != {} else 0).value_counts()
```

```python
list_of_production_companies = list(train['production_companies'].values)

Counter([i for j in list_of_production_companies for i in j]).most_common()
```

```python
top_production_companies = [str(m[0]) for m in Counter([i for j in list_of_genres for i in j]).most_common(30)]
for p in top_production_companies:
    train['production_companies_' + p] = train.production_companies.apply(lambda x: 1 if int(p) in x else 0)
    test['production_companies_' + p] = test.production_companies.apply(lambda x: 1 if int(p) in x else 0)

train['num_of_production_companies'] = train.production_companies.map(lambda x: len(x))
test['num_of_production_companies'] = test.production_companies.map(lambda x: len(x))
print(train.iloc[0])
```

```python
# production countries
# production_countires: US OR NONUS
#train.production_countries = train.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)]))
#test.production_countries = test.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)]))
#train['production_countries'].apply(lambda x: len(x) if x != {} else 0).value_counts()
train.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)])).apply(lambda x: len(x) if x != {} else 0).value_counts()

list_of_production_countries = list(train.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)])).values)
#print(list_of_production_countries)
Counter([i for j in list_of_production_countries for i in j]).most_common()
top_production_countries = [m[0] for m in Counter([i for j in list_of_production_countries for i in j]).most_common(25)]
print(top_production_countries)
```

```python
train.production_countries = train.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)])).map(lambda x: ','.join(map(str, x)))
test.production_countries = test.production_countries.map(lambda x: sorted([d['iso_3166_1'] for d in get_dictionary(x)])).map(lambda x: ','.join(map(str, x)))

for p in top_production_countries:
    train['production_countries_' + p] = train.production_countries.apply(lambda x: 1 if p in x else 0)
    test['production_countries_' + p] = test.production_countries.apply(lambda x: 1 if p in x else 0)
train['num_of_production_countries'] = train.production_countries.map(lambda x: len(get_dictionary(x)))
test['num_of_production_countries'] = test.production_countries.map(lambda x: len(x))

train['is_us_product'] = train.production_countries.apply(lambda x: 1 if 'US' in x else 0)
test['is_us_product'] = test.production_countries.apply(lambda x: 1 if 'US' in x else 0)
```

```python
## US OR non-US && multilanguage?
train.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)])).apply(lambda x: len(x) if x != {} else 0).value_counts()
list_of_spoken_languages = list(train.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)])).values)
top_spoken_languages = [str(m[0]) for m in Counter([i for j in list_of_spoken_languages for i in j]).most_common(15)]
#train.spoken_languages = train.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)]))
#test.spoken_languages = test.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)]))
```

```python
# spoken_languages
train.spoken_languages = train.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)])).map(lambda x: ','.join(map(str, x)))
test.spoken_languages = test.spoken_languages.map(lambda x: sorted([d['iso_639_1'] for d in get_dictionary(x)])).map(lambda x: ','.join(map(str, x)))
train['is_en_spoken'] = train.spoken_languages.apply(lambda x: 1 if 'en' in x else 0)
test['is_en_spoken'] = test.spoken_languages.apply(lambda x: 1 if 'en' in x else 0)

for p in top_spoken_languages:
    train['spoken_languages_' + p] = train.spoken_languages.apply(lambda x: 1 if p in x else 0)
    test['spoken_languages_' + p] = test.spoken_languages.apply(lambda x: 1 if p in x else 0)
train['num_of_spoken_languages'] = train.spoken_languages.map(lambda x: len(x))
test['num_of_spoken_languages'] = test.spoken_languages.map(lambda x: len(x))
```

```python
## TOP5
train.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)])).apply(lambda x: len(x) if x != {} else 0).value_counts()
list_of_keywords = list(train.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)])).values)
Counter([i for j in list_of_keywords for i in j]).most_common()
top_Keywords = [str(m[0]) for m in Counter([i for j in list_of_keywords for i in j]).most_common(30)]

#train.Keywords = train.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
#test.Keywords = test.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
```

```python
# top keyword
train.Keywords = train.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
test.Keywords = test.Keywords.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))

for p in top_Keywords:
    train['keyword_' + p] = train.Keywords.apply(lambda x: 1 if int(p) in x else 0)
    test['keyword_' + p] = test.Keywords.apply(lambda x: 1 if int(p) in x else 0)
train['num_of_keywords'] = train.Keywords.map(lambda x: len(x))
test['num_of_keywords'] = test.Keywords.map(lambda x: len(x))

#print(train.iloc[0])
```

```python
train.cast.map(lambda x: sorted([d['id'] for d in get_dictionary(x)])).apply(lambda x: len(x) if x != {} else 0).value_counts()
list_of_casts = list(train.cast.map(lambda x: sorted([d['id'] for d in get_dictionary(x)])).values)
Counter([i for j in list_of_casts for i in j]).most_common()
top_casts = [str(m[0]) for m in Counter([i for j in list_of_casts for i in j]).most_common(15)]

# cast
train.cast = train.cast.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
test.cast = test.cast.map(lambda x: sorted([d['id'] for d in get_dictionary(x)]))
train['num_of_cast'] = train.cast.map(lambda x: len(x))
test['num_of_cast'] = test.cast.map(lambda x: len(x))

for p in top_casts:
    train['cast_' + p] = train.cast.apply(lambda x: 1 if int(p) in x else 0)
    test['cast_' + p] = test.cast.apply(lambda x: 1 if int(p) in x else 0)

# crew
train.crew = train.crew.map(lambda x: len(get_dictionary(x)))
test.crew = test.crew.map(lambda x: len(get_dictionary(x)))
```

```python
for col in ['title', 'tagline', 'overview', 'original_title']:
    train['len_' + col] = train[col].fillna('').apply(lambda x: len(str(x)))
    test['len_' + col] = test[col].fillna('').apply(lambda x: len(str(x)))
```

```python
train.drop(['cast', 'status', 'production_countries', 'spoken_languages', 'Keywords', 'genres', 'production_companies', 'original_title', 'title', 'tagline', 'overview', 'homepage', 'imdb_id', 'poster_path'], inplace=True, axis = 1)
test.drop(['cast', 'status', 'production_countries', 'spoken_languages', 'Keywords', 'genres', 'production_companies', 'original_title', 'title', 'tagline', 'overview', 'homepage', 'imdb_id', 'poster_path'], inplace=True, axis = 1)
```

```python

train[['release_month','release_day','release_year']]=train['release_date'].str.split('/',expand=True).replace(np.nan, 0).astype(int)
test[['release_month','release_day','release_year']]=test['release_date'].str.split('/',expand=True).replace(np.nan, 0).astype(int)

train.loc[ (train['release_year'] <= 18) & (train['release_year'] < 100), "release_year"] += 2000
train.loc[ (train['release_year'] > 18)  & (train['release_year'] < 100), "release_year"] += 1900

test.loc[ (test['release_year'] <= 18) & (test['release_year'] < 100), "release_year"] += 2000
test.loc[ (test['release_year'] > 18)  & (test['release_year'] < 100), "release_year"] += 1900

list_of_release_year = list(train.release_year.map(lambda x: x).values)
#Counter([j for j in list_of_release_year]).most_common()
```

```python
train['_budget_year_ratio'] = (train['budget']/(train['release_year']*train['release_year']))
test['_budget_year_ratio'] = (test['budget']/(test['release_year']*test['release_year']))
```

```python
train_target = train.revenue
train_target = np.log1p(train_target)
train.drop(['id','revenue'], inplace=True, axis = 1)
print(train.iloc[0])
```

```python
from sklearn import *

#for c in ['original_language', 'status', 'genres', 'production_companies', 'production_countries', 'spoken_languages', 'Keywords']:
for c in ['original_language', 'belongs_to_collection']:
    le = preprocessing.LabelEncoder()
    le.fit(list(train[c].fillna('').astype(str)) + list(test[c].fillna('').astype(str)))
    train[c] = le.transform(train[c].fillna(''))
    test[c] = le.transform(test[c].fillna(''))

print(train.iloc[0])
print(train.iloc[1])

train.info()
test.info()
```

```python
from catboost import CatBoostRegressor
import xgboost as xgb
import lightgbm as lgb
from sklearn.model_selection import KFold
from sklearn.cross_validation import train_test_split
from datetime import datetime
%matplotlib inline

num_round = 100000
col = [c for c in train.columns if c not in ['id', 'revenue']]
cat_boost_model = CatBoostRegressor(
    iterations=num_round,
    learning_rate=0.003,
    depth=5,
    eval_metric='RMSE',
    colsample_bylevel=0.7,
    random_seed = 2019,
    bagging_temperature = 0.2,
    metric_period = None,
    early_stopping_rounds=500
)

k = 10
#fold = list(KFold(k, shuffle = True, random_state = 1).split(train))

#cat_test = test
#cat_test.revenue = cat_test.revenue.apply(lambda x: 0)
test_pred = np.zeros(test.shape[0])
#for i, (trn, val) in enumerate(fold):

for i in range(k):
    print(i+1, "fold.    RMSE")
    fold_test_pred = []
    X_train, X_test, y_train, y_test = train_test_split(train[col].fillna(-1), train_target, test_size = 0.3, random_state = i)
    #X_train, X_test, y_train, y_test = train.loc[trn, :], y[trn], train.loc[val, :], y[val]
    
    ## CatBoost
    cat_model = cat_boost_model.fit(
        X_train.values, y_train,
        eval_set = (X_test.values, y_test),
        use_best_model = True,
        verbose = False,
    )
    #val_pred = model.predict(val_x)
    fold_test_pred.append(cat_model.predict(test[col].fillna(-1).values))
    
    ## XGBoost
    xgb_params = {'learning_rate': 0.003, 'objective': 'reg:linear', 'max_depth': 6, 'subsample': 0.6, 'colsample_bytree': 0.7,  'eval_metric': 'rmse', 'seed': 2019, 'silent': True}
    dtrain = xgb.DMatrix(X_train.values, y_train, feature_names=col)
    dvalidate = xgb.DMatrix(X_test.values, y_test, feature_names=col)

    xgb_record = dict()
    bst = xgb.train(
        xgb_params,
        dtrain,
        num_round,
        [(dtrain, 'train'), (dvalidate, 'valid')],
        verbose_eval=False,
        callbacks = [xgb.callback.record_evaluation(xgb_record)],
        early_stopping_rounds=500
    )
    dtest = xgb.DMatrix(test[col].fillna(-1).values, feature_names=col)
    fold_test_pred.append(bst.predict(dtest, ntree_limit=bst.best_ntree_limit))
    xgb_best_idx = np.argmin(np.array(xgb_record['valid']['rmse']))
    
    ## LGB
    lgb_record = dict()
    lgb_params = {'objective':'regression', 'num_leaves' : 40, 'min_data_in_leaf' : 20, 'max_depth' : 4, 'learning_rate': 0.003, "feature_fraction": 0.8, "bagging_freq": 1, "bagging_fraction": 0.8, "bagging_seed": i, "metric": 'rmse', "random_state": 2019, "verbosity": -1}
    model = lgb.train(lgb_params
                      , lgb.Dataset(X_train.values, y_train)
                      , num_boost_round = num_round
                      , valid_sets = [lgb.Dataset(X_test.values, y_test)]
                      , early_stopping_rounds = 500
                      , callbacks = [lgb.record_evaluation(lgb_record)]
                      , verbose_eval=False
                     )
    fold_test_pred.append(model.predict(test, num_iteration = model.best_iteration))
    lgb_best_idx = np.argmin(np.array(lgb_record['valid_0']['rmse']))
    
    print("cat model.", "{0:.5f}".format(cat_model.get_best_score()['validation_0']['RMSE']))
    print("xgb model.", "{0:.5f}".format(xgb_record['valid']['rmse'][xgb_best_idx]))
    print("lgb model.", "{0:.5f}".format(lgb_record['valid_0']['rmse'][lgb_best_idx]))
    
    test_pred += (np.mean(np.array(fold_test_pred), axis = 0) / k)

#test.revenue = test_pred.apply(lambda x: x / k)
#test[['id', 'revenue']].to_csv('submission.csv', index=False)
#print(test.iloc[0])
```
